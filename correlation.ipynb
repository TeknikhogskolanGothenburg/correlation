```python
import pandas as pd

df = pd.read_csv('kc_house_data.csv')
```

```python
df.head()
```

```python
df.isnull().any()
```

```python
df[df['sqft_above'].isnull()]
```

```python
df['sqft_above'] = df['sqft_above'].fillna(0)
df[df['sqft_above'].isnull()]
```

```python
df.dtypes
```

```python
df.describe()
```

```python
from statsmodels.formula.api import ols

m = ols('price ~ bathrooms', df).fit()
print(m.summary())
```

```python
m = ols('price ~ sqft_living + grade + condition', df).fit()
print(m.summary())
```

```python
m = ols('price ~ sqft_living + grade + condition + waterfront', df).fit()
print(m.summary())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.jointplot(x="price", y="price", data=df, kind="reg", fit_reg=True, size=7)
plt.show()
```

```python
df.corr()
```

```python
import seaborn as sns
f,ax = plt.subplots(figsize=(25, 20))
sns.heatmap(df.corr(), annot=True, linewidths=1, fmt= '.1f',ax=ax)
plt.show()
```
